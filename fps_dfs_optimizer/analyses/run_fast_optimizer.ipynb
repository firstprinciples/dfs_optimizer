```python
import pandas as pd
from fps_dfs_optimizer.src.draftkings_tools import get_players_from_salaries
from fps_dfs_optimizer.src.generator import LineupGenerator
from fps_dfs_optimizer.src.draftkings_tools import EntriesHandler
```

```python
projection_file_name = '20211231_Night'

order = True #Set to none for random
n_lineups_to_optimize = 40
var_multiple = 1.0
time_limit = 4.0
team_limit_general = 4

batch_size = 20
drop_fraction = 0.0

projections_path = f'C:\\Users\\Jack\\Desktop\\proj\\{projection_file_name}.csv'
entries_output_path = f'..\\data\\entries\\{projection_file_name}_lineups.csv'

print(entries_output_path)
```

```python
df_players = get_players_from_salaries(projections_path)
df_players.head()
```

```python
team_limit = {t: team_limit_general for t in df_players['TeamAbbrev'].unique()}
```

```python
gen = LineupGenerator(
    df_players,
    batch_size=batch_size, 
    drop_fraction=drop_fraction, 
    duplicates_lim=100, 
    time_limit=time_limit, 
    verbose=False
)
```

```python
gen.generate_n_lineups(n_lineups_to_optimize, order=order, team_limits=team_limit)
```

```python
entries = EntriesHandler(None, gen.df)
entries.add_lineups_to_entries(gen.df_lineups, drop_entries=True, path=entries_output_path)
```
