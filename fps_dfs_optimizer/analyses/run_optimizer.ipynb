```python
import pandas as pd
from fps_dfs_optimizer.src.draftkings_tools import get_players_from_salaries
from fps_dfs_optimizer.src.generator import JacksOptimizer
from fps_dfs_optimizer.src.draftkings_tools import EntriesHandler
```

```python
projections_file_name = '20220107_Night'

n_lineups_to_optimize = 20
var_multiple = 1.0
use_auto_exposures = False

projections_path = f'C:\\Users\\Jack\\Desktop\\proj\\{projections_file_name}.csv'
generated_lineups_path = f'..\\data\\temp\\{projections_file_name}_gen.csv'
entries_output_path = f'..\\data\\entries\\{projections_file_name}_entries.csv'
```

```python
df_players = get_players_from_salaries(projections_path)
opt = JacksOptimizer(df_players, 20)
opt.df_lineups = pd.read_csv(generated_lineups_path)
```

```python
if use_auto_exposures:
    opt.enforce_exposures_auto(var_multiple=var_multiple, cov_penalty=0.002, n_lineups_to_optimize=n_lineups_to_optimize)
else:
    opt.enforce_exposures(var_multiple=var_multiple, n_lineups_to_optimize=n_lineups_to_optimize)

opt.get_player_distribution(opt.df_optimal)
```

```python
entries = EntriesHandler(None, opt.df)
entries.add_lineups_to_entries(opt.df_optimal, drop_entries=True, path=entries_output_path)
```
